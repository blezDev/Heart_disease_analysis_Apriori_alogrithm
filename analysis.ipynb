```python
agex = data['age'].mean()
def aged(x):
    if x > agex:
        return 1
    else:
        return 0
data['age'] = data['age'].apply(aged)
```

```python
tbx = data['trestbps'].mean()
def bp(x):
    if x > tbx:
        return 1
    else:
        return 0
data['trestbps'] = data['trestbps'].apply(bp)
```

```python
cholx = data['chol'].mean()
def cho(x):
    if x > cholx:
        return 1
    else:
        return 0
data['chol'] = data['chol'].apply(cho)
```

```python
thalx = data['thalach'].mean()
def tha(x):
    if x > thalx:
        return 1
    else:
        return 0
data['thalach'] = data['thalach'].apply(tha)
```

```python
oldx = data['oldpeak'].mean()
def old(x):
    if x > oldx:
        return 1
    else:
        return 0
data['oldpeak'] = data['oldpeak'].apply(old)
```

```python
def chp(x):
    if x > 2:
        return 1
    else:
        return 0
data['cp'] = data['cp'].apply(chp)
```

```python
def rest(x):
    if x == 0:
        return 0
    else:
        return 1
data['restecg'] = data['restecg'].apply(rest)
```

```python
def slp(x):
    if x >= 2:
        return 1
    else:
        return 0
data['slope'] = data['slope'].apply(slp)
```

```python
def ca(x):
    if x > 0:
        return 1
    else:
        return 0
data['ca'] = data['ca'].apply(ca)
```

```python
def th(x):
    if x > 3:
        return 1
    else:
        return 0
data['thal'] = data['thal'].apply(th)
```

```python
def cla(x):
    if x >= 1:
        return 1
    else:
        return 0
data['class'] = data['class'].apply(cla)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
freq = apriori(data, min_support=0.2, use_colnames=True, verbose=1)
freq
```

```python
rules = association_rules(freq, metric="lift", min_threshold=1)
rules
```
